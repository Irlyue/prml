```python
import kmeans
import numpy as np
import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
def load_data(n=50):
    x1 = np.random.randn(n, 2) * 0.8
    x2 = np.random.randn(n, 2) * 0.6 + np.array([1.5, 2.4])
    X = np.vstack((x1, x2))
    y = np.zeros(n*2, dtype=np.int32)
    y[n:] = 1
    return X, y
```

```python
np.random.seed(2)
X, y = load_data()
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.axis('equal')
```

```python
clr = kmeans.KMeans(2, n_steps=10)
```

```python
label, center = clr.fit(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=label)
plt.axis('equal')
```
